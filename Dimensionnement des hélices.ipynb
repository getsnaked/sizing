## Critères de dimensionnement des hélices 

Nous allons prendre nos hélices au nombre de 4 : une à chaque bout de bras du drone. Pour cela, nous pensons que le critère le plus important des hélices seraient qu'elles aient un bon rendement ce qui va donc directement influencer leur diamètre (aussi appélé pas par les constructeurs et généralement donné en pouce). Les hélices devront être capable de maintenir le drone en vol stationnaire mais c'est surtout durant la phase de décollage que les hélices devront fournir le plus de puissance et donc atteindront leur vitesse maximale. Au niveau des matériaux des héciles, nous pensons les prendre en plastique mais tout dépendra notamment de la vitesse à laquelle elles devront tourner. Comme nous souhaitons que notre drone est 4 hélices : nous allons donc en positionner 2 qui tourne dans le sens horaire et les 2 autres tourneront donc dans le sens inverse. En effet, si nos hélices tournent dans le même sens le drone va donc tourner sur lui même et ne pas pouvoir voler. De plus, il faudra également vérifier que les hélices et le moteur soient compatibles. 


### Equations pour les hélices

Une hélice en rotation va générer une force de traction ainsi que un couple qui va venir s'opposer au couple du moteur. Cette vitesse de rotation va se stabiliser lorsque le couple de l'hélice sera le même que le couple fourni par le moteur. 
En appliquant la formule de Newton on obtient l'équation de la poussée des hélices : 
$T = C_t \times \rho \times n^2$
